```python
import seaborn as sns
df = sns.load_dataset("car_crashes")

numerical_df = df.select_dtypes(include=['number'])

corr_matrix = numerical_df.corr()
sns.heatmap(corr_matrix, annot=True)
```

From the data I can tell
1. Alchol was the leading cause of car crashes
2. Nearly 70% of intoxicated drivers were speeding when they crash.
3. Almost all the crashes reported states that it was the driver's first accident.
